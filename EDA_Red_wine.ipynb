### Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in predicting the quality of wine.

### Key Features of the Red Wine Quality Data Set and Their Importance

The Red Wine Quality Data Set contains various physicochemical properties of red wine samples, along with a quality rating. Here are the key features and their importance in predicting wine quality:

1. **Fixed Acidity**: This refers to the non-volatile acids that do not evaporate during fermentation. It contributes to the wine's tartness and stability.

2. **Volatile Acidity**: High levels of volatile acidity can lead to unpleasant vinegar-like taste. It is critical for determining wine spoilage.

3. **Citric Acid**: Adds freshness and flavor to the wine. Small amounts can add 'freshness' and reduce spoilage.

4. **Residual Sugar**: The amount of sugar remaining after fermentation stops. It can affect sweetness and the perception of dryness in the wine.

5. **Chlorides**: Represents the salt content in wine. High levels can result in a salty taste which is generally undesirable.

6. **Free Sulfur Dioxide**: The free form of SO2, which prevents microbial growth and oxidation. Important for wine preservation.

7. **Total Sulfur Dioxide**: The total amount of SO2 in its bound and free form. Excess can lead to a noticeable chemical taste.

8. **Density**: Influenced by sugar and alcohol content. It helps in inferring the alcohol and sugar level of the wine.

9. **pH**: Measures the acidity of the wine. It affects the taste, stability, and color.

10. **Sulphates**: Added to prevent oxidation and to increase shelf life. They can affect the aroma and taste.

11. **Alcohol**: Impacts the body, flavor, and perception of sweetness. Higher alcohol content can enhance certain flavors and affect mouthfeel.

12. **Quality**: The target variable that represents the sensory score given by wine experts. This is a subjective measure but is influenced by the other physicochemical properties.

### Importance in Predicting Wine Quality

- **Fixed Acidity**: Important as it impacts the wine's overall stability and acidity profile.
- **Volatile Acidity**: Crucial for quality control as high levels can spoil the wine.
- **Citric Acid**: Enhances flavor and stability, thus affecting quality positively.
- **Residual Sugar**: Influences sweetness, a critical factor in consumer preference.
- **Chlorides**: Essential for detecting undesired salty taste, impacting quality negatively.
- **Free and Total Sulfur Dioxide**: Vital for preserving wine, preventing spoilage, and oxidation.
- **Density**: Indirectly related to alcohol and sugar content, affecting mouthfeel and body.
- **pH**: Affects taste, stability, and color, making it significant for overall quality.
- **Sulphates**: Important for preservation and taste profile.
- **Alcohol**: Directly impacts the flavor profile and consumer preference.

Understanding the interplay of these features is key to predicting and improving wine quality, as they collectively influence the sensory characteristics that define a high-quality wine.


### Q2. How did you handle missing data in the wine quality data set during the feature engineering process? Discuss the advantages and disadvantages of different imputation techniques.

### Handling Missing Data in the Wine Quality Data Set

During the feature engineering process, handling missing data is crucial to ensure the quality and reliability of the analysis. Here are common techniques for dealing with missing data and their advantages and disadvantages:

### Imputation Techniques

### 1. **Mean/Median Imputation**
Replacing missing values with the mean or median of the respective feature.

- **Advantages**:
  - Simple and quick to implement.
  - Retains the overall distribution of the data.

- **Disadvantages**:
  - Can introduce bias, especially if the data is not normally distributed.
  - Ignores the relationship between features.

### 2. **Mode Imputation**
Replacing missing values with the mode (most frequent value) of the respective feature.

- **Advantages**:
  - Useful for categorical variables.
  - Simple to implement.

- **Disadvantages**:
  - Can introduce bias if the mode is not representative of the missing values.
  - Less effective for continuous variables.

### 3. **K-Nearest Neighbors (KNN) Imputation**
Using the k-nearest neighbors to estimate the missing values based on the similarity of other features.

- **Advantages**:
  - Takes into account the relationships between features.
  - Often more accurate than mean/median imputation.

- **Disadvantages**:
  - Computationally intensive, especially for large datasets.
  - Requires careful tuning of the parameter \( k \).

### 4. **Multivariate Imputation by Chained Equations (MICE)**
Using a series of regression models to estimate missing values, considering the relationships between all features.

- **Advantages**:
  - Provides robust estimates by considering multivariate relationships.
  - Suitable for datasets with complex interactions.

- **Disadvantages**:
  - Computationally intensive and complex to implement.
  - Requires assumptions about the data distribution.

### 5. **Interpolation and Extrapolation**
Using mathematical interpolation (e.g., linear, polynomial) to estimate missing values.

- **Advantages**:
  - Effective for time series and data with a logical order.
  - Retains the trends and patterns in the data.

- **Disadvantages**:
  - Can be inaccurate if the data does not follow a clear trend.
  - Not suitable for categorical data.

### 6. **Dropping Missing Values**
Removing rows or columns with missing values.

- **Advantages**:
  - Simple and easy to implement.
  - No bias introduced by imputation.

- **Disadvantages**:
  - Can lead to significant loss of data, especially if missing values are prevalent.
  - Reduces the sample size, potentially impacting the statistical power of the analysis.

## Choosing the Right Technique

The choice of imputation technique depends on the nature of the data and the extent of missingness. For the wine quality data set, the following considerations apply:

- **For small amounts of missing data**: Mean/median or mode imputation may be sufficient.
- **For data with complex relationships**: KNN or MICE imputation would be more appropriate.
- **For time series data**: Interpolation techniques can be effective.
- **When missing data is substantial**: Dropping rows/columns may not be feasible, making imputation necessary.

Properly handling missing data ensures the integrity of the dataset and the reliability of subsequent analyses and model predictions.

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the matplotlib figure
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 10))
fig.suptitle('Distribution of Each Feature in the Wine Quality data Set')

# List of features
features = df.columns

# Plot distribution for each feature
for i, ax in enumerate(axes.flatten()):
    if i < len(features):
        sns.histplot(df[features[i]], kde=True, ax=ax)
        ax.set_title(features[i])
    else:
        ax.axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
````

````python
from scipy.stats import shapiro

# Perform Shapiro-Wilk test for normality
normality_results = {}
for feature in features:
    stat, p = shapiro(df[feature])
    normality_results[feature] = (stat, p)

# Display features with p-value < 0.05 (indicating non-normality)
non_normal_features = {k: v for k, v in normality_results.items() if v[1] < 0.05}
non_normal_features
````

# 1. Log Transformation

````python
# Apply log transformation
data_log_transformed = df.copy()
for feature in non_normal_features.keys():
    data_log_transformed[feature] = data_log_transformed[feature].apply(lambda x: np.log(x + 1))

# Plot transformed distributions
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 10))
fig.suptitle('Distribution of Log Transformed Features')

for i, ax in enumerate(axes.flatten()):
    if i < len(non_normal_features):
        sns.histplot(data_log_transformed[list(non_normal_features.keys())[i]], kde=True, ax=ax)
        ax.set_title(list(non_normal_features.keys())[i])
    else:
        ax.axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
````

# 2. Square Root Transformation


````python
# Apply square root transformation
data_sqrt_transformed = df.copy()
for feature in non_normal_features.keys():
    data_sqrt_transformed[feature] = data_sqrt_transformed[feature].apply(lambda x: np.sqrt(x))

# Plot transformed distributions
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 10))
fig.suptitle('Distribution of Square Root Transformed Features')

for i, ax in enumerate(axes.flatten()):
    if i < len(non_normal_features):
        sns.histplot(data_sqrt_transformed[list(non_normal_features.keys())[i]], kde=True, ax=ax)
        ax.set_title(list(non_normal_features.keys())[i])
    else:
        ax.axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
````

# 3. Box-Cox Transformation


````python
from scipy.stats import boxcox

# Apply Box-Cox transformation
data_boxcox_transformed = df.copy()
for feature in non_normal_features.keys():
    # Box-Cox transformation requires positive data
    data_boxcox_transformed[feature] = data_boxcox_transformed[feature].apply(lambda x: x + 1)
    data_boxcox_transformed[feature], _ = boxcox(data_boxcox_transformed[feature])

# Plot transformed distributions
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 10))
fig.suptitle('Distribution of Box-Cox Transformed Features')

for i, ax in enumerate(axes.flatten()):
    if i < len(non_normal_features):
        sns.histplot(data_boxcox_transformed[list(non_normal_features.keys())[i]], kde=True, ax=ax)
        ax.set_title(list(non_normal_features.keys())[i])
    else:
        ax.axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
````

### Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

## Principal Component Analysis (PCA) on the Wine Quality Data Set

### Step 1: Import Libraries and Load Data

````python


import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
# Load the wine quality data set

df=pd.read_csv(r"C:\Users\Lenovo\Desktop\New Notebook\Notes\ML\winequality-red.csv",sep=';')

# Display the first few rows of the data set
df.head()
````

To perform Principal Component Analysis (PCA) on the wine quality dataset and determine the minimum number of principal components required to explain 90% of the variance, follow these steps:

1. **Load the dataset**: Load the wine quality dataset.
2. **Standardize the data**: Standardize the features to have mean = 0 and variance = 1.
3. **Apply PCA**: Perform PCA on the standardized data.
4. **Calculate explained variance**: Calculate the cumulative explained variance to determine how many principal components are required to explain 90% of the variance.

Here's the implementation:



````python


# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Apply PCA
pca = PCA()
pca.fit(scaled_data)

# Calculate explained variance
explained_variance = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance.cumsum()

# Determine the number of components required to explain 90% of the variance
num_components = (cumulative_explained_variance >= 0.90).argmax() + 1

# Plot the cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(cumulative_explained_variance, marker='o', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Principal Components')
plt.axhline(y=0.90, color='r', linestyle='-')
plt.axvline(x=num_components, color='r', linestyle='-')
plt.grid(True)
plt.show()


num_components
````
